# Amino acid substitution ranking

Define the function to quantify the difference between the two amino acids

Define the variables needed for the ranking

```python
def aasub(first, second):
    
    aa1lettercode = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']
    aa3lettercode = ['ala', 'cys', 'asp', 'glu', 'phe', 'gly', 'his', 'ile', 'lys', 'leu', 'met', 'asn', 'pro', 'gln', 'arg', 'ser', 'thr', 'val', 'trp', 'tyr']
    
    
    
    if len(first) == 3:
        
        
    
    
    
    size = ['Tiny', 'Tiny', 'Small', 'Average', 'Average', 'Tiny', 'Average', 'Average', 'Average', 'Average', 'Average', 'Small', 'Small', 'Average', 'Average', 'Tiny', 'Small', 'Small', 'Average', 'Average']
    polarity = ['Non-polar', 'Polar', 'Polar', 'Polar', 'Non-polar', 'Non-polar', 'Polar', 'Non-polar', 'Polar', 'Non-polar', 'Non-polar', 'Polar', 'Non-polar', 'Polar', 'Polar', 'Polar', 'Polar', 'Non-polar', 'Polar', 'Polar']
    water = ['Hydrophobic', 'Hydrophillic', 'Hydrophillic', 'Hydrophillic', 'Hydrophobic', 'Hydrophobic', 'Hydrophillic', 'Hydrophobic', 'Amphipathic', 'Hydrophobic', 'Amphipathic', 'Hydrophillic', 'Hydrophobic', 'Hydrophillic', 'Hydrophillic', 'Hydrophillic', 'Hydrophillic', 'Hydrophobic', 'Amphipathic', 'Amphipathic']
    structure = ['Aliphatic', 'Sulphur-containing', 'Acidic', 'Acidic', 'Aromatic', 'Simple', 'Basic', 'Aliphatic', 'Basic', 'Aliphatic', 'Sulphur-containing', 'Amidic', 'Cyclic', 'Amidic', 'Basic', 'Hydroxylic', 'Hydroxylic', 'Aliphatic', 'Aromatic', 'Aromatic']
    nucleophillic = ['No', 'Yes', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'No', 'No', 'No', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes']
    PTM1 = ['None', 'None', 'None', 'None', 'None', 'None', 'None', 'None', 'Commonly methylated', 'None', 'None', 'None', 'None', 'None', 'Commonly methylated', 'Commonly phosphorylated', 'Commonly phosphorylated', 'None', 'None', 'Commonly phosphorylated']
    PTM2 = ['None', 'None', 'Commonly Glycosylated', 'None', 'None', 'None', 'None', 'None', 'Commonly Ubiquitinated', 'None', 'None', 'None', 'None', 'None', 'None', 'Commonly Glycosylated', 'Commonly Glycosylated', 'None', 'None', 'None']
    PKACOOH = [2.35, 1.71, 2.09, 2.19, 1.83, 2.34, 1.82, 2.36, 2.18, 2.36, 2.28, 2.02, 1.99, 2.17, 2.18, 2.21, 2.63, 2.32, 2.38, 2.2]
    PKANH2 = [9.69, 10.78, 9.82, 9.67, 9.13, 9.6, 9.17, 9.68, 8.95, 9.68, 9.21, 8.84, 10.6, 9.13, 9.04, 9.15, 9.1, 9.62, 9.39, 9.11]
    PKASC = [0, 8.33, 3.86, 4.25, 0, 0, 6.04, 0, 10.79, 0, 0, 0, 0, 0, 12.48, 0, 0, 0, 0, 10.07]
    IEP = [6.02, 9.56, 2.98, 3.22, 5.48, 5.97, 7.61, 6.02, 9.87, 6.02, 5.75, 5.43, 6.3, 5.65, 10.76, 5.68, 5.87, 5.97, 5.89, 9.59]
    MM = [89, 121, 132, 146, 165, 75, 155, 131, 147, 131, 149, 132, 115, 146, 175, 105, 119, 117, 204, 181]

    traits = [polarity, structure, nucleophillic, PTM1, PTM2]
    traitstr = ['Polarity', 'Structure', 'Nucleophillic', 'PTM1', 'PTM2']

    firstloc = aa1lettercode.index(first)
    secondloc = aa1lettercode.index(second)

    difference = 0
    differenttraits = []

    # Polarity, Structure, Nucleophillic, PTM1, PTM2

    counter = 0
    for item in traits:
        if item[firstloc] != item[secondloc]:
            difference = difference + 1
            differenttraits.append(traitstr[counter])
            counter = counter + 1
        else:
            counter = counter + 1

    # Water
    if water[firstloc] == water[secondloc]:
        difference = difference + 0
    elif water[firstloc] == 'Amphipathic' or water[secondloc] == 'Amphipathic':
        difference = difference + 0.5
        differenttraits.append('Half Water')
    else:
        difference = difference + 1
        differenttraits.append('Water')

    # Size
    if size[firstloc] == size[secondloc]:
        difference = difference + 0
    elif size[firstloc] == 'Tiny':
        if size[secondloc] == 'Small':
            difference = difference + 0.5
            differenttraits.append('Half Size')
        else:
            difference = difference + 1
            differenttraits.append('Size')
    elif size[secondloc] == 'Tiny':
        if size[firstloc] == 'Small':
            difference = difference + 0.5
            differenttraits.append('Half Size')
        else:
            difference = difference + 1
            differenttraits.append('Size')
    else:
        difference = difference + 1
        differenttraits.append('Size')

    # PKA COOH
    COOHdif = abs(PKACOOH[firstloc]-PKACOOH[secondloc])

    if COOHdif < 0.3:
        difference = difference + 0
    elif COOHdif < 0.6:
        difference = difference + 0.5
        differenttraits.append('Half COOHdif')
    else:
        difference = difference + 1
        differenttraits.append('COOHdif')

    # PKA NH2
    NH2dif = abs(PKANH2[firstloc]-PKANH2[secondloc])

    if NH2dif < 0.64:
        difference = difference + 0
    elif NH2dif < 1.28:
        difference = difference + 0.5
        differenttraits.append('Half NH2dif')
    else:
        difference = difference + 1
        differenttraits.append('NH2dif')

    # PKA SC
    SCdif = abs(PKASC[firstloc]-PKASC[secondloc])

    if SCdif < 2.87:
        difference = difference + 0
    elif SCdif < 5.75:
        difference = difference + 0.5
        differenttraits.append('Half SCdif')
    else:
        difference = difference + 1
        differenttraits.append('SCdif')

    # IEP
    IEPdif = abs(IEP[firstloc]-IEP[secondloc])

    if IEPdif < 2.59:
        difference = difference + 0
    elif IEPdif < 5.19:
        difference = difference + 0.5
        differenttraits.append('Half IEPdif')
    else:
        difference = difference + 1
        differenttraits.append('IEPdif')

    # MM
    MMdif = abs(MM[firstloc]-MM[secondloc])

    if MMdif < 43:
        difference = difference + 0
    elif MMdif < 86:
        difference = difference + 0.5
        differenttraits.append('Half MMdif')
    else:
        difference = difference + 1
        differenttraits.append('MMdif')
        
    # Summary
    print("\n\nThe difference value between the amino acid {0} and {1} is: {2}".format(first, second, difference))
    print("The traits which differ are: {0}".format(', '.join(differenttraits)))

```
